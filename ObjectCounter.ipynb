Libraries


```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Reading image and conversion to grayscale

```python
image = cv2.imread('coins.jpeg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

Making image blur to avoid noises

```python
blur = cv2.GaussianBlur(gray, (15, 15), 0)
plt.imshow(blur, cmap='gray')
```

Detecting edges using canny algorithm

```python
canny = cv2.Canny(blur, 30, 100, 5)
plt.imshow(canny, cmap='gray')
```

Connecting edges for better visibility

```python
dilated = cv2.dilate(canny, (5, 5), iterations=1)
plt.imshow(dilated, cmap='gray')
```

Calculating and drawing contours

```python
(cnt, hierarchy) = cv2.findContours(
	dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)

plt.imshow(rgb)
```

Result

```python
print("coins in the image : ", len(cnt))
```
